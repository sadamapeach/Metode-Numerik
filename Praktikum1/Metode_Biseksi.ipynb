```python
# Program biasa
import numpy as np
import warnings
import sympy

def bisection(f,a,b,tol):
  MAX_STEPS = 1000
 
  # Initialize
  delta_x = b - a
  c = a + delta_x / 2.0
  c_array = [c]
 
  f_a = f(a)
  f_b = f(b)
  f_c = f(c)

  # Header judul
  print('\t\t\t\t\tPencarian Akar Fungsi f(x) = e**x + x**2 - 3x -2 = 0')
  print('\t\t\t\t\t\t\tDengan Metode Biseksi')
  print('=================================================================================================================================')

  print("""
---------------------------------------------------------------------------------------------------------------------------------
| Iter\t|       xl      |       xr\t|       xm\t|       delta      |       f(xl)      |       f(xm)\t| f(x1)*f(xm)\t|
---------------------------------------------------------------------------------------------------------------------------------""")

  # Periksa kurung
  if np.sign(f_a) == np.sign(f_b):
    raise ValueError("tidak mengurung: f(a) dan f(b) harus memiliki tanda yang berbeda")
 
  # Loop sampai kita mencapai TOLERANSI atau kita mengambil MAX_STEPS
  for step in range(1, MAX_STEPS + 1):
  # Periksa toleransi â€“ bisa juga memeriksa ukuran delta_x
  # Kita memeriksa ini terlebih dahulu karena kita atelah menginisiasi nilai-nilai dalam c dan f_c
    print('|  {}\t| {:.10f}\t| {:.10f}\t|  {:.8f}\t|   {:.10f}   |   {:.10f}   |  {:.10f}\t| {:.10f}\t|'.format(step,a,b,c,delta_x,f_a,f_c,f_a*f_c))
    if np.abs(f_c) < tol:
      break
    if np.sign(f_a) != np.sign(f_c):
      b = c
      f_b = f_c
    else:
      a = c
      f_a = f_c
    delta_x = b - a
    c = a + delta_x / 2.0
    f_c = f(c)
    c_array.append(c)

  print('---------------------------------------------------------------------------------------------------------------------------------')

  if step == MAX_STEPS:
    warnings.warn('Jumlah maksimum langkah yang terlampaui')
 
  return c, np.array(c_array)
```

```python
# mengatur fungsi sebagai fungsi lambda inline
f = lambda x: sympy.exp(x) + x**2 - 3*x - 2
# interval awal
a = -1.0
b = 1.0
tol = 1.e-5
#Mencari akar
x_star, x_array = bisection(f, a, b, tol)
print('\nAkar pendekatan\t:', x_star)
print('Toleransi\t: 0.00001')
```
