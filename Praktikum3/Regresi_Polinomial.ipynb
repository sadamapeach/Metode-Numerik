```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

```python
# Regresi linear kuadrat terkecil menggunakan optimize.curve_fit from scipy
from scipy.optimize import curve_fit
x = np.array([0.05, 0.15, 0.46, 0.70, 0.82, 1.17])
y = np.array([0.956, 0.832, 0.571, 0.378, 0.306, 0.104])

def func(x, a, b, c):
  y = a + b*x + c*(x**2)
  return y

alpha, covs = curve_fit(func, xdata = x, ydata = y)
print(alpha)
```

```python
# Memasukkan nilai alpha yang sudah didapat ke
# persamaan polinomial orde dua y = a1 + a2.x + a3.x^2
f = lambda x: alpha[0] + alpha[1]*x + alpha[2]*(x**2)

# Plot the results
plt.figure(figsize = (8,6))
plt.plot(x, y, 'r.')
plt.plot(x, f(x), 'b')
plt.title('Model Regresi Polinomial Orde Dua')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
