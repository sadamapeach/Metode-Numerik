```python
import numpy as np
import matplotlib.pyplot as plt # menggambarkan grafik
plt.style.use('seaborn-poster')
%matplotlib inline

# Definisi parameter
f = lambda t, s: t + s # ODE
# t = x dan s = y

# Metode Euler 
def euler(f,x0,y0,xn,n):
  h = (xn-x0)/n # ukuran langkah 
  print('h = ', h)
  t = np.arange(x0, xn + h, h) # grid Numerik
  s0 = y0 # kondisi awal
  
  # Metode Euler
  s = np.zeros(len(t))
  s[0] = s0
  for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])
    t[i + 1] = t[i] + h
  return t, s

x0 = 0
y0 = 1
xn = 0.1
n  = 10 # n = (xn-x0)/h

# Menampilkan grafik
t,s = euler(f,x0,y0,xn,n)
plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Aproksimasi')
plt.plot(t, np.exp(t) - t - 1, 'g', label='Exact') 
plt.title('Solusi Aproksimasi dan Exact untuk ODE sederhana')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()

# Perhitungan galat
sol_euler = s[10]
sol_exact = np.exp(0.1) - 0.1 - 1
galat_ra = abs((sol_exact-sol_euler)/sol_exact)*100

print('ya =',sol_euler)
print('ys =',sol_exact)
print('galat relatif absolut = ',galat_ra,'%')
```
