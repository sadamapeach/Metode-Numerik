```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

```python
x = [321.0, 322.8, 324.2, 325.0]
y = [2.50651, 2.50893, 2.51081, 2.51188]

# P1_coeff = [-1476207.552, -13668.69618, 42.1875, -0.043402778]
# P2_coeff = [-6100679.113, 56594.19192, -175, 0.18037518]
# P3_coeff = [9396236.049, -87291.79688, 270.3125, -0.279017857]
# P4_coeff = [-4771763.49, 44366.30114, -137.5, 0.142045455]

P1_coeff = [(850295550/576), -(7873169/576), (24300/576), -(25/576)]
P2_coeff = [-(4227770625/693), (39219775/693), -(121275/693), (125/693)]
P3_coeff = [(4209513750/448), -(39106725/448), (121100/448), -(125/448)]
P4_coeff = [-(839830374/176), (7808469/176), -(24200/176), (25/176)]

# get the polynomial function
P1 = poly.Polynomial(P1_coeff)
P2 = poly.Polynomial(P2_coeff)
P3 = poly.Polynomial(P3_coeff)
P4 = poly.Polynomial(P4_coeff)

x_new = np.arange(320, 326, 0.1) 

fig = plt.figure(figsize = (12,8))
plt.plot(x_new, P1(x_new), 'b', label = 'P1')
plt.plot(x_new, P2(x_new), 'r', label = 'P2')
plt.plot(x_new, P3(x_new), 'g', label = 'P3')
plt.plot(x_new, P4(x_new), 'y', label = 'P4')

plt.plot(x, np.ones(len(x)), 'ko', x, np.zeros(len(x)), 'ko')
plt.title('Lagrange Basis Polynomials')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

```python
L = 2.50651*P1 + 2.50893*P2 + 2.51081*P3 + 2.51188*P4
print('L(x) = ',L)
print('\n')

fig = plt.figure(figsize = (10,8))
plt.plot(x_new, L(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()

def lagrange_interpolation(x, y, xx):
  n = len(x)
  sum = 0
  for i in range(n):
    product = y[i]
    for j in range(n):
      if i != j:
        product = product*(xx - x[j])/(x[i]-x[j])
    sum = sum + product
  return sum
print('Hasil')
print(lagrange_interpolation([321.0, 322.8, 324.2, 325.0], [2.50651, 2.50893, 2.51081, 2.51188], 323.5))
```
